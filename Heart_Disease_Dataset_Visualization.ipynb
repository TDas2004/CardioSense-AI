# Heart Disease Prediction System Analysis (EDA)

## Dataset Overview

The dataset contains clinical and demographic attributes of patients used to analyze the presence of heart disease. It includes both numerical and categorical variables along with a binary target variable indicating disease status.

## Objective

To perform exploratory data analysis (EDA) on a heart disease dataset to understand the distribution of clinical and demographic features and their relationship with heart disease occurrence.

To identify key risk factors associated with heart disease using statistical analysis and data visualizations.

To derive meaningful insights that can support early diagnosis and data-driven decision-making in healthcare.

## Tools Used


    Python
    Pandas
    Numpy
    Matplotlib
    Seaborn


## Code Cell

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"final_joined_dataset.csv")
df
```

## Target Variable Analysis

```python
ax = sns.countplot( x = 'target', data = df, hue = 'sex', palette = "Set2")

plt.xticks(ticks=[0, 1],labels=["No Heart Disease", "Heart Disease"])


ax.legend(title="Gender",labels=["Female", "Male"])

plt.ylabel("Number of Patients")
plt.title('Target Variable Analysis')

plt.show()
```

**Insight** The dataset contains more male patients than female patients in both heart disease and non–heart disease groups, indicating a gender imbalance in the sample.

## Univariate Analysis 

### Numerical columns

### AGE

```python
sns.histplot( x = 'age', data = df, bins = 30,  kde = True)

plt.xlabel('Age')
plt.title('Age Distribution of Patients')

plt.show()
```

**Insight** Age is approximately normally distributed, with most patients concentrated between middle age, indicating the dataset is dominated by middle-aged individuals

### RestingBP

```python
sns.histplot( x = 'trestbps', data = df, bins = 20, kde = True, stat = 'density')

plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.title('Distribution of Resting Blood Pressure (trestbps)')

plt.show()
```

**Insight** Resting blood pressure is slightly right-skewed, with most patients clustered in the normal to mildly elevated range and a few high-value outliers.

## Cholesterol

```python
sns.histplot(x = 'chol', data = df, kde = True)

plt.xlabel('Cholesterol')
plt.title('Distribution of Cholesterol Levels')

plt.show()
```

**Insight** Cholesterol levels are right-skewed, with most patients clustered in the borderline-to-high range and a few extreme high-value outliers.

```python
sns.boxplot(x = 'sex', y = 'chol', data = df, hue = 'sex')

plt.xticks(ticks = [0,1], labels = ["Female","Male"])
    
plt.xlabel("Gender")
plt.ylabel("Cholesterol")
plt.title("Cholesterol Distribution by Gender")

plt.show()
```

**Insight** Female patients show a slightly higher median cholesterol level than male patients, with overlapping distributions and outliers present in both groups.

## Max Heart Rate

```python
sns.histplot( x = 'thalach', data = df, kde = True)

plt.xlabel('Max Heart Rate(during intense work)')
plt.title('Distribution of Maximum Heart Rate Achieved (thalach)')

plt.show()
```

**Insight** Most patients achieved maximum heart rates within the expected exercise-induced range, indicating normal cardiovascular response during stress testing.

```python
sns.boxplot(x="target", y="thalach", data=df, hue = 'target')


plt.xticks(ticks=[0, 1],labels=["No Heart Disease", "Heart Disease"])

plt.xlabel('Heart Disease Status')
plt.ylabel('Max Heart Rate')
plt.title("Maximum Heart Rate by Heart Disease Status")

plt.show()
```

**Insight** Patients with heart disease tend to show lower maximum heart rate compared to those without the disease.

## OldPeak (heart stress response)

```python
sns.histplot( x = 'oldpeak', data = df, kde = True)

plt.xlabel('Heart Stress Response (Higher oldpeak = worse heart stress response)')
plt.title('Heart Stress Response')

plt.show()
```

**Insight** Oldpeak values are heavily right-skewed, with most patients showing low ST depression, while a small subset exhibits high values indicating increased cardiac stress.

```python
sns.boxplot(x="target", y="oldpeak", data=df, hue = 'target')

plt.xticks(ticks=[0, 1],labels=["No Heart Disease", "Heart Disease"])

plt.xlabel("Heart Disease Status")
plt.ylabel("Oldpeak (ST Depression)")
plt.title("Oldpeak by Heart Disease Status")

plt.show()
```

**Insight** Patients with heart disease exhibit higher median oldpeak values and greater variability compared to those without heart disease, indicating stronger exercise-induced ST depression.

### Categorical columns

### Gender

```python
sns.countplot(x = 'sex', data = df, hue = 'sex', palette = 'Set2', legend = False)

plt.xticks(ticks = [0,1], labels = ['Female', 'Male'])

plt.xlabel("")
plt.title("Gender Distribution of Patients")

plt.show()
```

**Insight** The dataset is male-dominated, with a substantially higher number of male patients compared to female patients

### ChestPainType

```python
sns.countplot(x = 'cp', data = df, hue = 'sex', palette = 'Set2', legend = False)

plt.xticks(ticks = [0,1,2,3,4], labels = ['Typical angina', 'Atypical angina', 'Non-anginal pain', 'Asymptomatic','Asymptomatic'])

plt.xticks(rotation = 45, ha = "right")


plt.title("Distribution of Chest Pain Types")
plt.xlabel("Chest Pain Type")

plt.show()
```

**Insight** Asymptomatic chest pain is the most common category across both genders, particularly among male patients, highlighting the prevalence of silent cardiac symptoms in the dataset.

### ExerciseAngina

```python
sns.countplot(x="exang", data=df, hue = 'exang', palette = 'Set2', legend = False)

plt.xticks([0, 1], ["No", "Yes"])

plt.xlabel("")
plt.title("Exercise-Induced Angina Distribution")

plt.show()
```

**Insight** Most patients do not experience exercise-induced angina, while a substantial minority report angina during physical activity, indicating varying exercise tolerance within the dataset.

### ST_Slope

```python
sns.countplot(x = "slope", data = df, hue = 'slope', palette = 'Set1')

plt.xticks([0,1,2,3], ["Upsloping", "Flat", "Downsloping","Downsloping"])

plt.title("ST Segment Slope Distribution")

plt.show()
```

**Insight** Flat and downsloping ST segments dominate the dataset, indicating a high prevalence of abnormal ECG responses during exercise.

## Bivariate Analysis

### Age vs Target

```python
sns.boxplot( x = 'target', y = 'age', data = df, hue = 'target', legend = False)

plt.xticks([0, 1], ["No Heart Disease", "Heart Disease"])

plt.xlabel("")
plt.ylabel("Age")
plt.title("Age Distribution by Heart Disease Status")

plt.show()
```

**Insight** Patients with heart disease tend to be slightly older on average, with a higher median age compared to those without heart disease, though the age distributions overlap considerably

### Cholesterol vs Target

```python
sns.boxplot( x = 'target', y = 'chol', data = df, hue = 'target', legend = False)

plt.xticks([0, 1], ["No Heart Disease", "Heart Disease"])

plt.xlabel("")
plt.ylabel("Cholesterol")
plt.title("Cholestrol Level by Heart Disease Status")

plt.show()
```

**Insight** Cholesterol levels show a slightly higher median in patients with heart disease, with considerable overlap between the two groups and the presence of high-value outliers in both.

## Multivariate / Relationship Analysis

### Age vs Max HR

```python
plt.figure(figsize=(10, 6))

sns.scatterplot( data = df, x = "age", y = "thalach", hue = "target", alpha = 0.7, s = 25, edgecolor = 'white')

plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.title('Age vs Maximum Heart Rate by Heart Disease Status')

plt.show()
```

**Insight** Maximum heart rate decreases with increasing age, and patients with heart disease generally achieve lower maximum heart rates compared to those without heart disease, though there is overlap between the groups.

### Age vs Cholesterol

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data = df, x = "age", y = "chol", hue = "target",palette = 'Set1', alpha = 0.7, s = 25, edgecolor = 'white')

plt.xlabel("Age")
plt.ylabel("Cholesterol (mg/dL)")
plt.title("Age vs Cholesterol by Heart Disease Status")

plt.show()
```

**Insight** There is no strong linear relationship between age and cholesterol levels, as values are widely scattered across all ages, with substantial overlap between patients with and without heart disease.

## Correlation Analysis

```python
plt.figure(figsize=(12, 8))

corr = df.select_dtypes(include="number").corr()

sns.heatmap(corr, cmap = "coolwarm", center = 0, annot = False)

plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

**Insight** Maximum heart rate (thalach) shows a strong negative correlation with age, while oldpeak exhibits a positive correlation with heart disease, indicating its potential importance as a predictive feature.

## Key Insights

- Patients with heart disease tend to be slightly older and achieve lower maximum heart rates, indicating reduced exercise capacity.

- Oldpeak (ST depression) shows a clear increase in heart disease cases, making it one of the most informative features.

- Exercise-related variables (maximum heart rate, ST slope, exercise-induced angina) provide stronger signals of heart disease than cholesterol alone.

- Cholesterol levels show high variability across all ages and disease statuses, with no strong standalone relationship to heart disease.

- The dataset is male-dominated, and asymptomatic chest pain is common, highlighting the presence of silent cardiac risk.
